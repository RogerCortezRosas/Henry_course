## `Series`

```python
# Como crear una serie de una lista , un arreglo numpy y un diccionario

lista = list('abcdefghijklmnopqrstuvwxyz')
array = np.arange(26)
diccionario = dict(zip(lista,array))

ser1 = pd.Series(lista)
ser2 = pd.Series(array)
ser3 = pd.Series(diccionario)

#print(ser1)
#print(ser2)
print(ser3)
```

```python
# Cómo convertir el índice de una serie en una columna de un dataframe

lista = list('abcdefghijklmnopqrstuvwxyz')
array = np.arange(26)
diccionario = dict(zip(lista,array))

ser = pd.Series(diccionario)

df = ser.to_frame().reset_index()

print(df)
```

```python
# Series como diccionarios especializados

population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}

Serie_population = pd.Series(population_dict)

print(Serie_population)  # Impresion de toda la serie
print()
print(Serie_population.values) # Impresion de los valores de la serie
print()
print(Serie_population.index) # Impresion de los indices de una serie
print()
print(Serie_population.values[1:4]) # Impresion de uno mas valores en especifico de acuerdo a su posicion
print()
print(Serie_population.index[0:3]) # Impresion de uno o mas indices de acuerdo a su posicion
print()
print(Serie_population['Texas'])# Impresion del valor de un indice
print()
print(Serie_population['California':'New York']) # Impresion slicing desde el indice California hasta New york
```

```python
# Un escalar repetido a lo largo de un índice
pd.Series(5, index = [100, 200, 300])
```

```python
# Como obtener los elementos de la serie A que no estan presentes en B

A = pd.Series([1, 2, 3, 4, 5])
B = pd.Series([4, 0, 6, 7, 8])

A[~A.isin(B)]
```

```python
#Como obtener los elementos que no tienene en comun la serie A y la serie B

ser1 = pd.Series([1, 2, 3, 4, 5])
ser2 = pd.Series([4, 5, 6, 7, 8])

ser_u = pd.Series(np.union1d(ser1,ser2)) # union
ser_i = pd.Series(np.intersect1d(ser1,ser2))#intersect
ser_u[~ser_u.isin(ser_i)] # el simbolo ~ determina la negacion del algoritmo en ese caso todos aquellos elementos que no coinciden
```

```python
# Como obtener el minimo , 25th percentile,medina,75th y el maximo de una serie numerica

state = np.random.RandomState(100)
ser2 = pd.Series([100,100 , 100, 100, 100])

percent = np.percentile(ser2,q=[0,25,50,75,100])
percent
```

### `Index` como un set ordenado

* Se pueden utilizar operaciones de conjuntos con los ``Index`` siguiendo las convenciones de Python

```python
indA = pd.Index([1, 3, 5, 7, 9])
indB = pd.Index([2, 3, 5, 7, 11])

indC = indA & indB # interseccion
print(indC)
print()

indC = indA | indB # union
print(indC)
print()

indC = indA ^ indB  # symmetric difference
print(indC)
print()
```

The describe() method returns description of the data in the DataFrame.

If the DataFrame contains numerical data, the description contains these information for each column:

count - The number of not-empty values.
mean - The average (mean) value.
std - The standard deviation.
min - the minimum value.
25% - The 25% percentile*.
50% - The 50% percentile*.
75% - The 75% percentile*.
max - the maximum value.

```python
# Concatencaion de data Frames
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
   

df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])


df3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                    'B': ['B8', 'B9', 'B10', 'B11'],
                    'C': ['C8', 'C9', 'C10', 'C11']},
                    index=[8, 9, 10, 11])

frames = [df1, df2, df3]
df3
```
